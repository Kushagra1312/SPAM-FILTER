##                                    IMPORTING LIBRARIES

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# IMPORTING DATASETS

```python
dataset = pd.read_csv('emails.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X
```

IMPORTING DATASET CONTAINING SPAM WORDS.

```python
spamdataset = pd.read_csv('spamwords.csv')
spam = spamdataset.iloc[:,-1].values
l=[]
for word in spam:
    l.append(word.lower())
            
l   
```

## MAKING THE FINAL LIST TO WORK WITH LOGISTIC REGRESSION

```python
l2=[]
for sentences in X:
    s = str(sentences)
    count = 0
    for spam in l:
        if spam in s:
            count = count+1
    l2.append(count/len(s))
l2 
```

## SPLITTING  DATASET 


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(l2, y, test_size = 0.25,random_state=1)
```

```python
X_train = np.array(X_train)
X_train = X_train.reshape(len(X_train),1)
X_train
```

## RESHAPING X_TEST

```python
X_test = np.array(X_test)
X_test = X_test.reshape(len(X_test),1)
X_test
```

## APPLYING FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## APPLYING LOGISTIC REGRESSION MODEL


```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state =1)
classifier.fit(X_train, y_train)
```

## PREDICTING THE TEST RESULTS

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## CHECKING THE ACCURACY OF THE MODEL


```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
